```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://www.moscow_city.vybory.izbirkom.ru/region/region/moscow_city?action=show&root=1&tvd=27720002197406&vrn=27720002197402&region=77&global=&sub_region=77&prver=0&pronetvd=null&vibid=27720002197406&type=222"
```

```python
response = requests.get(url)
response.encoding = 'cp1251'
htmlString = response.text
page = BeautifulSoup(htmlString, "html.parser")
```

```python
dump_folder = 'tmp'
```

```python
 print(page.prettify())
```

```python
table_inner_titles = page.find('table', align="left", cellpadding="2", cellspacing="1", vspace="0")
rows = table_inner_titles.find_all('tr')
#print (len(rows))
#print (table_inner_titles.prettify())
for rows in table_inner_titles.find_all('tr'):
    cols = table_inner_titles.find_all('td')

titles = []
for place in cols[3:35:3]:
    titles.append(place.text)
titles

total_number = []
for place in cols[4:36:3]:
    total_number.append(place.text[:len(place.text)-2])
total_number

candidates = []
for place in cols[38::3]:
    candidates.append(place.text)
candidates

cand_votes = []
cand_votes_percent = []
for place in cols[39::3]:
    stop = place.text.index('\n')
    cand_votes.append(place.text[:stop])
```

```python
table_all_nums = page.find('table', align="left", bgcolor="#ffffff", border="0", cellpadding="2", cellspacing="1", style="width:100%;overflow:scroll", vspace="0")
#print (table_all_nums.prettify())
rows = table_all_nums.find_all('tr')
for row in table_all_nums.find_all('tr'):
    cols = row.find_all('a')
links = rows[0].find_all('a')
districts = []
for link in links:
    districts.append(link.text)
districts
```

```python
numbers_dict = dict.fromkeys(titles)
for num, title in enumerate(numbers_dict):
    all_votes = rows[num + 1].find_all('b')
    numbers_dict[title] = [total_number[num]] + [all_vote.text for all_vote in all_votes]
numbers_dict
```

```python
cands_dict = dict.fromkeys(candidates)
for num, place in enumerate(cands_dict):
    all_votes = rows[num + 13].find_all('b')
    cands_dict[place] = [cand_votes[num]] + [all_vote.text for all_vote in all_votes]
```

```python
#cands_dict
#disctricts
#numbers_dict
df_cands = pd.DataFrame(cands_dict)
df_cands
df_votes = pd.DataFrame(numbers_dict)
df_votes
```

```python
district = ['Total'] + districts
df_votes['District'] = district
df_votes
```

```python
df_cands.to_csv(os.path.join(dump_folder, ('candidates.csv')), sep=';')
df_votes.to_csv(os.path.join(dump_folder, ('votes.csv')). seps=';')
```
